```python
words = ["eat","tea","tan","ate","nat","bat"]
word_dict = {}
for word in words:
    key = ''.join(sorted(word))
    if key not in word_dict:
        word_dict[key] = []
    word_dict[key].append(word)
print(word_dict)
```

```python
nums = [1,5,7,-1,5]
res = []
unique_set = set()
target = 6
for i in nums:
    complement = target - i
    if complement not in unique_set:
        res.append((complement,i))
        unique_set.add(complement)
        unique_set.add(i)
print(res)
```

```python
lst = [1,[2,3],[4,[5,6]]]
res = []
def flatten(lst):

    for i in lst:
        if isinstance(i,list):
            flatten(i)
        else:
            res.append(i)
    return res
```

```python
l1 = [1,2,4]
l2 = [1,3,4]

i, j = 0, 0
res = []

while i < len(l1) and j < len(l2):
    if l1[i] <= l2[j]:
        res.append(l1[i])
        i += 1
    else:
        res.append(l2[j])
        j += 1

# Add remaining elements
res.extend(l1[i:])
res.extend(l2[j:])

print(res)
```
