```python
s = "aabbcde"
count_char = {}
for c in s:
    count_char[c] = count_char.get(c,0)+1
print(count_char)
```

```python
arr = [1,2,4,2,3,3,4,3]
num_dict = {}
for num in arr:
    if num not in num_dict:
       num_dict.setdefault(num,0)
    num_dict[num] += 1
print(num_dict)
```

```python
def two_sums(nums,target):
    lookup = {}
    for i,num in enumerate(nums):
        diff = target - num
        if diff in lookup:
            return [lookup[diff],i]
        lookup[num] = i

print(two_sums([2,7,11,15], 9))
```
