```python
#Calculate the factorial of a number entered by the user
fact = 1
n = int(input("Enter any number"))
for i in range(1,n+1)[::-1]:
    fact *= i
print("factorial of {} is {}".format(n,fact))
```

```python
#Search for a number in a list. Stop the loop once found.
list1 = [13,45,67,24,38]
find_num = 67
index = None
for i in range(len(list1)):
    if list1[i] == find_num:
        index = i
print("{} found at {}".format(find_num,index))
```

```python
#Print characters of a string but skip vowels
str1 = "hello I am learning Object Oriented programming Language"
filtered = []
for i in str1 :
    if i in "aieou":
        continue
    filtered.append(i)
print("".join(i for i in filtered))
```

```python
#Print all prime numbers between 1 and 500 using nested loops.
for i in range(1,500):
    flag = 0
    if i == 2:
        print("2 is prime number")
    elif i > 2:
        for k in range(2,i):
            if i % k == 0:
                flag = 1
                continue
        if not flag :
            print(f"{i} is prime number")
    else:
        print("Neither prime nor composite")
        
```

```python
str1 = "programming"

for ch in str1:
    count = 0
    for c in str1:
        if ch == c:
            count += 1
    print(ch, "=", count)
```

```python
#pattern 1
for i in range(6):
    for j in range(i):
        print("*",end='')
    print()
```

```python
#pattern 2
n = 5
for i in range(n):
    for j in range(1,n-i+1):
        print("*",end='')
    print()
```

```python
#Count how many numbers between 1–100 contain digit 5.
count = 0
for i in range(1,101):
    num = str(i)
    if '5' in num:
        count += 1
print(count)
```
