```python
current_choice = ""
booked_movies = []
available_movies = ["Avengers", "Inception", "Interstellar", "Joker", "Titanic"]
valid_choices = [str(i+1) for i in range(len(available_movies))]

while current_choice != '0':
    print("Choose a movie to book")
    for index,movie in enumerate(available_movies):
        print(index +1," : ",movie)
    
    current_choice = input("Enter your choice: ")
    if current_choice in valid_choices:
        booked_movies.append(available_movies[int(current_choice)-1])
        print("Movie has been booked {}".format(available_movies[int(current_choice)-1]))
    elif current_choice == '0':
        break
    else:
        print("Invalid choices")
print("Your booked movies:", booked_movies)
```

```python
current_choice = ""
booked_movies = []
available_movies = ["Avengers", "Inception", "Interstellar", "Joker", "Titanic"]
valid_choices = [str(i+1) for i in range(len(available_movies))]

while current_choice != '0':
    print("Choose a movie to book")
    for index,movie in enumerate(available_movies):
        print(index +1," : ",movie)
    
    current_choice = input("Enter your choice: ")
    if current_choice in valid_choices:
        index = int(current_choice)-1
        chosen_movie = available_movies[index]
        if chosen_movie in booked_movies:
           print("Removing {}".format(chosen_movie))
           booked_movies.remove(chosen_movie)
        else:
           print("Booking {}".format(chosen_movie))
           booked_movies.append(chosen_movie)
        print("Movie has been booked {}".format(available_movies[int(current_choice)-1]))
    elif current_choice == '0':
        break
    else:
        print("Invalid choices")
print("Your booked movies:", booked_movies)
```
