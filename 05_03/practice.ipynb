```python
arr = [1, 2, 3, 4, 5]
# Output: [5, 4, 3, 2, 1]
arr1 = [i for i in arr[::-1]]
print(arr1)
```

```python
arr = [10, 20, 30]
sum = 0
for i in arr:
    sum += i
print(sum)
```

```python
#count even numbers in list
count = 0
arr = [1,2,3,4,5,6]
for i in arr:
    if i % 2 == 0:
        count += 1
print(count)
```

```python
arr = ['a', 'b', 'c']
for index, value in enumerate(arr):
    print(index, value)
```

```python
arr = [1,2,2,3,4,3,5]
#Remove duplicates from list
arr.sort()
num = 0
for i in reversed(arr):
    if i == num:
        arr.remove(i)
    else:
        num = i

print(arr)
```

```python
#Find second largest element
arr = [10, 20, 4, 45, 99]
print(sorted(arr,reverse=True)[1:2])
```

```python
arr = [(1,3), (4,1), (2,2)]
print(sorted(arr,key=lambda x:x[1]))
```

```python
arr = [1,2,3,4,5]
#square even numbers only
sq_even = [i*i for i in arr if i % 2 == 0]
print(sq_even)
```

```python
arr = [1,2,2,3,3,3]

freq = {}

for num in arr:
    freq[num] = freq.get(num, 0) + 1

print(freq)
```

```python
arr = [[1,2], [3,4], [5]]
#flatten list
# arr1 = [i for i in j for j in arr]
arr1 = []
for i in arr:
    for j in i:
        arr1.append(j)
print(arr1)
```

```python
a = [1,2,3,4]
b = [3,4,5,6]
for i in a:
    for j in b:
        if i == j:
            print(i)
```

```python
arr = [1,2,3,4,5]
k = 2
#rotate list in k steps
k = k %len(arr)
rotated = arr[-k:]+arr[:-k]
print(rotated)
```
