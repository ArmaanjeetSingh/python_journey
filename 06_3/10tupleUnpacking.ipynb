```python
#Use tuple unpacking inside a loop.
students = [("Aman", 78), ("Riya", 65), ("Karan", 82), ("Simran", 90)]
for detail in students:
    name,marks = detail
    print("{} scored {} marks".format(name,marks))
```

```python
#@Print only students who scored 40 or more.
for detail in students:
    name,marks = detail
    if marks > 40:
        print("{} passed".format(name))
```

```python
students = [("Aman", 78), ("Riya", 65), ("Karan", 82), ("Simran", 90)]
# Use enumerate() to print ranking starting from 1
rank_list = sorted(students,key = lambda x:x[1],reverse=True)
for rank,(name,marks) in enumerate(rank_list,start=1):
    print("Rank", rank, ":", name, "-", marks)
```

```python
products = [("Laptop", 55000), ("Mouse", 500), ("Keyboard", 1200)]
#tuple unpacking.
for index,(name,price) in enumerate(products):
    print(f"Product {name} Price: {price}")
```

```python
temps = [("Delhi", 35), ("Mumbai", 32), ("Shimla", 18), ("Chennai", 36)]
for index,(city,temp) in enumerate(temps):
    if temp > 34:
        print(city,end=',')
```

```python
students = [("Aman", 78), ("Riya", 65), ("Karan", 82)]

for index, student in enumerate(students):
    name, marks = student
    print(index, name, marks)
```
