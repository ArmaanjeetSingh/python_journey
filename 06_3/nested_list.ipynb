```python
orders = [
    ("Vinayak", "Burger", 120),
    ("Rahul", "Pizza", 250),
    ("Aman", "Burger", 120),
    ("Sonia", "Pasta", 200),
    ("Riya", "Pizza", 250)
] 
```

```python
#Print all customer names.
for name,item,price in orders:
    print(f"{name} : {item}-{price}")
```

```python
#Count how many Pizza orders were made.
count = 0
for name,item,price in orders:
    if item == "Pizza":
        count += 1
print(count)
```

```python
#Calculate the total revenue from all orders.
total_revenue = 0
for name,item,price in orders:
    total_revenue += price
print(total_revenue)
```

```python
#Print customers who ordered Burger.
for name,item,price in orders:
    if item.casefold() == "burger":
        print(name)
```

```python
#Find the most expensive item ordered
expensive_price = 0
expensive_item = ""
for name,item,price in orders:
    if expensive_price < price:
        expensive_price = price
        expensive_item = item
print(f"{expensive_item} has most expensive pricing {expensive_price}")
```

```python
students = [
    ("Aman", (78, 85, 90)),
    ("Riya", (92, 88, 79)),
    ("Rahul", (60, 70, 65)),
    ("Sneha", (88, 91, 95))
]
# Print each student's average marks.
for name,(m1,m2,m3) in students:
    print(f"{name} : {(m1+m2+m3)/3:.2f}")
```

```python
#Find the student with highest total marks.
std_nm = ""
std_marks = 0
for name,(m1,m2,m3) in students:
    total = m1 + m2 + m3
    if std_marks < total:
        std_marks = total
        std_nm = name
print(f"{name} has achieved highest total marks {std_marks:.2f}")
```

```python
#Print students whose average is above 85.
for name,(m1,m2,m3) in students:
    avg = (m1 + m2 + m3)/3
    if avg > 85:
        print(f"{name} has average {avg:.2f}")
```

```python
#Create a new list containing only student names.
student_names = []
for name,(m1,m2,m3) in students:
    student_names.append(name)
print(student_names)
```

```python
shopping_list = ["milk", "bread", "eggs", "rice", "pasta", "milk"]
flag = "" #for if item exists before in list
#Remove duplicate items
shopping_list.sort()
for index,value in enumerate(reversed(shopping_list)):
    if flag == value :
        del shopping_list[index]
    else:
        flag = value
print(shopping_list)
```

```python
#Count how many times milk appears
count = 0
for item in shopping_list:
    if item.casefold() == "milk":
        count += 1
print("Milk appears {} times".format(count))
```

```python
inventory = [
    ["Laptop", 10, 50000],
    ["Phone", 25, 20000],
    ["Tablet", 15, 30000],
    ["Headphones", 50, 2000]
]
#Print all product names
for name,qty,price in inventory:
    print(name,end=' ')
```

```python
#Find the total value of inventory
total_value = 0
for name,qty,price in inventory:
    total_value += (qty * price)
print(f"{total_value:32.20f}")
```

```python
#Print products with quantity less than 20
for name,qty,price in inventory:
    if qty < 20:
        print(f"{name} with quantity {qty}")
```

```python
#Increase price of Phone by 10%.
for name,qty,price in inventory:
    if name == "Phone":
       price +=  price *0.1
       print(f"{name} : {price}")
```

```python
movies = [
    ("Inception", [5,4,5,5,4]),
    ("Avatar", [4,4,3,5,4]),
    ("Interstellar", [5,5,5,4,5])
]
for name,(r1,r2,r3,r4,r5) in movies:
    avg = (r1+r2+r3+r4+r5) / 5
    print(f"{name} : {avg}")
```

```python
#Print movies with average rating above 4.5.
for name,(r1,r2,r3,r4,r5) in movies:
    avg = (r1+r2+r3+r4+r5) / 5
    if avg > 4.5:
        print(f"{name} : {avg}")
```
