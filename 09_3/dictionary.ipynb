```python
students = {
    "Rahul": 78,
    "Aman": 85,
    "Simran": 92,
    "Riya": 67
}
#print all students and marks 
for name,marks in students.items():
    print(f"{name} - {marks}")
```

```python
#Print students who scored above 80
for name,marks in students.items():
    if marks > 80:
        print(f"{name}")
```

```python
#Find highest marks
max_marks = 0
std_name = None
for name,marks in students.items():
    if marks > max_marks:
       max_marks = marks
       std_name = name
        
print(f"{std_name} - {max_marks}")
```

```python
menu = {
    "burger": 120,
    "pizza": 250,
    "pasta": 180,
    "coke": 50
}
# Show menu items with numbers
for index,value in enumerate(menu,start=1):
    print(f"{index} - {value}")
```

```python
valid_choices = {str(index):value for index,value in enumerate(menu,start=1)}
user_choice = input("Enter your choice")
if user_choice in valid_choices:
    print(f"{valid_choices[user_choice]} is selected")
    selected_item = valid_choices[user_choice]
    print(f"price : {menu[selected_item]}")
```

```python
inventory = {
    "apple": 10,
    "banana": 6,
    "milk": 5,
    "bread": 4
}
for item,qty in inventory.items():
    print(f"{item} : {qty}")
```

```python
#Ask user which item to buy
user_choice = input("Enter your item ").casefold()
required_qty = int(input("How many "))

if user_choice in inventory:

    if required_qty <= inventory[user_choice]:
        inventory[user_choice] -= required_qty
        print("Remaining {} : {}".format(user_choice, inventory[user_choice]))

    else:
        print("Not enough stock")

else:
    print("Item not found")
```

```python
products = {
    "laptop": 50000,
    "mouse": 500,
    "keyboard": 1500,
    "monitor": 12000
}
cart = {}
for product,price in products.items():
    print(f"{product} : {price}")
```

```python
valid_choices = {str(index):value for index,value in enumerate(products,start=1)}

while True:
    for index,val in valid_choices.items():
        print(index,'-',val)

    user_choice = input("Enter your product choice ")

    if user_choice in valid_choices:
       selected_product = valid_choices[user_choice]
       qty = int(input("Enter quantity"))
       cart[selected_product] = qty
       print("Added to cart {}".format(selected_product))

    if user_choice=='0':
        break

    
    

print(cart)
```

```python
print("Final bill")
for product,qty in cart.items():
    print(f"{product} X {qty} = {products[product]*qty}")
```
