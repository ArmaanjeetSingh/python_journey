```python
#Write a function that counts vowels in a string.
def count_vowels(text:str)->int:
    count  = 0
    for char in text:
        if char in 'aeiou':
            count += 1
    return count
count_vowels("python programming")
```

```python
#Write a function that returns the reversed string.
def reverse_string(text:str)->str:
    return text[::-1]
reverse_string("python")
```

```python
#Write a function that returns True if the number is even
def is_even(num:int)->bool:
    if num % 2 == 0:
        return True
    else:
        return False
is_even(8)
```

```python
#Write a function that joins list of words with space.
def join_words(words:list[str])->str:
    joined_str = ' '.join(words)
    return (joined_str)
join_words(["I", "love", "Python"])
```

```python
#Write a function that returns the largest number in a list
def find_max(num_list:list)->int:
    max_n = 0
    for i in range(len(num_list)):
        if max_n < num_list[i]:
            max_n = num_list[i]

    return max_n
find_max([3,7,2,9,1])
```

```python
#Convert words into uppercase
def to_uppercase(words: list[str]) -> list[str]:
    return_words = []
    for element in words:
        return_words.append(element.upper())
    return return_words

to_uppercase(["python","java"])
```
