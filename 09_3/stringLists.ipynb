```python
#Splitting into list
text = "Python is very powerful"
l = text.split()
print(l)
```

```python
#Joining into string
words = ["I", "love", "Python"]
print(' '.join(words))
```

```python
#Convert into comma separated 
data = "apple,banana,mango,grapes"
csv = data.split(',')
print(csv)
```

```python
#join using '-' 
items = ["2026", "03", "09"]
print('-'.join(items))
```

```python
#Count snumber of words
count  = 1
text = "Python programming is fun"
for item in text:
    if not item.isalpha():
        count +=1
print(count)
```

```python
#Reverse words in a sentence
text = "learn python step by step"
l = text.split()
print(' '.join(l[::-1]))
```

```python
#Extract file extension
file = "report.pdf"
file_split = file.split('.')
print(f"Extension : {file_split[1]}")
```

```python
#create username from email
email = "vinayak@gmail.com"
username = email.split("@")[0]
print(username)
```

```python
#convert to hashtag format
text = "learn python fast"
text2 = text.split()
print('#'.join([i for i in text2]))
```
