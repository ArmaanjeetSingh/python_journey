```python
arr = [1,2,2,3,4,4,5]
#find unique elements
unique = set()
for item in arr:
    if item not in unique:
        unique.add(item)
print(list(unique))
```

```python
a = (1,2)
#swap tuple values
a = list(a)
temp = a[1]
a[1] = a[0]
a[0] = temp
print(tuple(a))
```

```python
#Sort list of numbers
arr = [5,2,9,1]
for i in range(len(arr)-1):
    if arr[i] > arr[i+1]:
        temp = arr[i]
        arr[i] = arr[i+1]
        arr[i+1] = temp
print(arr)
```

```python
s = "aabbcde"
seen = []
duplicate = []
for char in s:
    if char not in seen:
        seen.append(char)
    else:
        duplicate.append(char)
for x in duplicate:
    seen.remove(x)
print(seen)
```

```python
d = {'a':[1,2], 'b':[3,4]}
#flatten dictionary values
l = [item for item in d.values()]
l1 = []
for i in l:
    for j in i:
        l1.append(j)
print(l1)
```

```python
arr = [1,2,2,3,3,3]
#find most frequent element
count_dict = {}
for i in arr:
    count_dict[i] = count_dict.get(i,0)+1
l = list(zip(count_dict.keys(),count_dict
.values()))
print(max(l,key = lambda x:x[1]))
```
