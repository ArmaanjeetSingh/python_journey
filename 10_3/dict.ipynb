```python
sentence = "apple banana apple mango banana apple"
words = sentence.split()
count = {}
for word in words:
    if word not in count:
        count[word] = 1
    else:
        count[word]+= 1
print(count)
```

```python
marks = {
    "Rahul":85,
    "Aman":40,
    "Simran":72,
    "Neha":33,
    "Riya":90
}
updated_marks = {}
for name,mark in marks.items() :
    if mark >= 50:
       updated_marks.setdefault(name,marks[name])
```

```python
print(updated_marks)
```

```python
prices = {
    "apple":100,
    "banana":40,
    "mango":120
}

new_prices = {}

for fruit, price in prices.items():
    new_prices[fruit] = price

print(new_prices)
```
