```python
library_books = {"python", "java", "c++", "dbms"}
def add_book(book:str)->None:
    library_books.add(book)

    print("{} is added".format(book))

def remove_book(book:str)->None:
    try :
       library_books.remove(book)
    except KeyError:
        print("book not available")
def borrow_book(book:str)->None:
    if book in library_books:
        library_books.remove(book)
        print(f"You borrowed {book}")
    else:
        print("Book not available")
add_book("ai")
borrow_book("python")
remove_book("dbms")
```

```python
student1 = {"python", "math", "ai", "dbms"}
student2 = {"python", "ml", "math", "cloud"}
#Common subjects both students study
print(student1 & student2) #intersection
print(student1 | student2) #union
print(student1 - student2)
print(student2 - student1)
```

```python
rahul_friends = {"aman", "deepak", "simran", "neha"}
aman_friends = {"simran", "neha", "rahul", "karan"}
mutual_friends = rahul_friends.intersection(aman_friends)

all_friends = rahul_friends.union(aman_friends)

unique_to_aman = aman_friends.symmetric_difference(rahul_friends)
unique_to_rahul = rahul_friends.symmetric_difference(aman_friends)
print(mutual_friends,all_friends,unique_to_aman,unique_to_rahul,sep='\n')
```

```python
packing_items = {
    "clothes",
    "phone",
    "knife",
    "laptop",
    "water bottle",
    "explosives"
}

plane_restricted = {"knife", "explosives"}
train_restricted = {"explosives"}
def check_luggage(mode:str)->None:
    if mode == "plane":
        allowed_items = packing_items-plane_restricted
        print(allowed_items)
    elif mode == "train":
        allowed_items = packing_items-train_restricted
        print(allowed_items)
    else:
        print("Invalid mode")
check_luggage("plane")
check_luggage("train")
```

```python
patient_drugs = {"aspirin", "metformin", "warfarin"}
dangerous_combo = {"warfarin", "ibuprofen"}
harmful =  patient_drugs& dangerous_combo
if harmful :
    for drug in harmful :
        print("Warning: Dangerous drug detected ",drug)
else:
    print("Not harmful")
```

```python
prescription = {"warfarin", "metformin", "amlodipine"}
contraindicated = {"warfarin"}
replacement = {"aspirin"}
# Remove contraindicated drugs
# Add safer replacement
for drug in contraindicated:
    if drug in prescription:
        prescription.remove(drug)
        print(f"{drug} removed from prescription")
    else:
        print("No such drug found")
prescription |= replacement
print(prescription)
```

```python
quote = """
Life is like riding a bicycle. To keep your balance you must keep moving
"""
prepositions = {"is","like","to","your","a"}
words = quote.split()
set_words = set(words)
print(set_words.intersection(prepositions))
```

```python
backend = {"python", "java", "nodejs", "go"}
frontend = {"javascript", "html", "css", "react", "nodejs"}
print("Technologies used in both ",backend | frontend)
print("Technologies used only in frontend ",frontend - backend)
print("Technologies used only backend ",backend - frontend)
```
