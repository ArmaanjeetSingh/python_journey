```python
list1 = [1,2,2,3,1,4,2]
# Given a list of numbers, count how many times each number appears.

dict1 = {}
for i in list1:
    if i not in dict1:
        dict1[i] = 1
    else:
        dict1[i] += 1
print(dict1)
```

```python
dict1.clear()
for i in list1:
    dict_val = dict1.get(i,0)+1
    dict1[i] = dict_val
print(dict1)
```

```python
l = [1,2,3,4,2,5,1,6]

seen = set()
duplicates = set()

for i in l:
    if i in seen:
        duplicates.add(i)
    else:
        seen.add(i)

print(duplicates)
```

```python
string1 = "python is easy and python is powerful"
#Given a sentence, remove duplicate words using a set.
l = string1.split()
l1 = []
seen = set()
for i in l:
    if i not in seen:
        seen.add(i)
        l1.append(i)
```

```python
print(' '.join(l1))
```

```python
# Swap keys and values.
dict1 = {"a":1,"b":2,"c":3}
dict_copy = dict1.copy()
dict1.clear()
for key,val in dict_copy.items():
    dict1[val] = key
print(dict1)
```

```python
#group words by first letter
list1  = ["apple","banana","apricot","blueberry","cherry"]
dict1 = {}
for word in list1:
    first_letter = word[0]
    dict1.setdefault(first_letter,[]).append(word)
print(dict1)
```

```python
str1 = "programming"
#Find the First Non-Repeating Character
dict1 = {}
for i in str1:
    dict1[i] = dict1.setdefault(i,0)+1
print(dict1)
print(list(zip(dict1,dict1.values())))
```

```python
list1 = [1,2,2,3,4,4,5]
#Return elements that appear only once.
seen = {}
for num in list1 :
    seen[num] = seen.setdefault(num,0)+1
for num in seen:
    if seen[num] == 1:
        print(num,end=',')
```

```python
freq = {}

for num in list1:
    freq[num] = freq.get(num,0) + 1

for num in freq:
    if freq[num] == 1:
        print(num, end=",")
```

```python
list1 = [1,2,3,4]
list2 = [2,3,5]
list3 = [2,3,6]
#Find Common Elements in Multiple Lists
set1 = set(list1)
set2 = set(list2)
set3 = set(list3)
print(set1 & set2 & set3)
```

```python
anagrams = ["eat","tea","tan","ate","nat","bat"]
grouped = {}
#Group Anagrams
for word in anagrams:
    key = ''.join(sorted(word))
    grouped.setdefault(key,[]).append(key)
print(grouped)
    
```

```python
details = {
"user":{
    "name":"Rahul",
    "age":22
},
"city":"Delhi"
}
new_details = {}
for key,val in details.items():
    if isinstance(val, dict):
        for key2,value2 in val.items():
            new_details[key + "." + key2] = value2

    else:
        new_details[key] = val

print(new_details)
```

```python
items = [
{"id":1,"name":"A"},
{"id":2,"name":"B"},
{"id":1,"name":"A"}
]
#Remove duplicate items
seen = set()
result = []

for d in items:
    if d["id"] not in seen:
        seen.add(d["id"])
        result.append(d)

print(result)
```

```python
items = [
{"id":1,"name":"A"},
{"id":2,"name":"B"},
{"id":1,"name":"A"}
]
#Remove duplicate items
seen = set()
result = []
for d in items:
    t = tuple(d.items())
    if t not in seen:
        seen.add(t)
        result.append(d)
print(result)
```

```python
nums = [2,7,11,15]
target = 9
#two sum
for i in range(len(nums)):
    for j in range(i+1,len(nums)):
        if nums[i] + nums[j] == target:
            print(nums[i],nums[j])
```
