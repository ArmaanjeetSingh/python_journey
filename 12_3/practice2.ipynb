```python
data = [
("Rahul", 85),
("Simran", 90),
("Rahul", 85),
("Aman", 78),
("Simran", 90)
]#Remove duplicate tuples and print the unique records.
result = []
seen = set()
for name,marks in data:
    if name not in seen:
        seen.add(name)
        result.append((name,marks))
for detail in result:
    print(detail)
```

```python
locations = {
(28.61, 77.20): "Delhi",
(19.07, 72.87): "Mumbai",
(12.97, 77.59): "Bangalore"
}

def find_city(coord:tuple)->str:
    city = locations.get(coord,None)
    if city :
        print(city)
```

```python
students = {
"Rahul": {"Python", "DBMS", "AI"},
"Simran": {"Python", "Java"},
"Aman": {"Python", "AI"}
}
common_sub = None
# Find subjects that all students study.
for name, detail in students.items():
    if common_sub is None:
        common_sub = detail
    else:
        common_sub &= detail

print(common_sub)
    
```

```python
employees = {
101: ("Rahul", "HR"),
102: ("Simran", "IT"),
103: ("Aman", "IT"),
104: ("Neha", "HR"),
105: ("Ravi", "Finance")
}
job_posts = set()
for key,val in employees.items():
    if val[1] not in job_posts:
        job_posts.add(val[1])

```

```python
orders = [
("Rahul", "Laptop"),
("Rahul", "Mouse"),
("Simran", "Laptop"),
("Aman", "Keyboard"),
("Simran", "Mouse"),
("Rahul", "Laptop")
]
orders_dict = dict()
for name,product in orders:
    if name in orders_dict:
        orders_dict[name].append(product)
    else:
        orders_dict[name] = [product]
print(orders_dict)
```

```python
orders_dict = dict()
for name,product in orders:
    orders_dict.setdefault(name,set()).add(product)
```

```python
print(orders_dict)
```

```python
#Create a dictionary showing number of items bought by each person.
orders_count = dict()
for name,product in orders:
        orders_count[name] = orders_count.get(name,0)+1
print(orders_count)
```

```python
data = [
(1, 2),
(2, 3),
(3, 4),
(1, 4)
]
unique_nums = set()

for num1, num2 in data:
    unique_nums |= {num1, num2}

print(unique_nums)
```

```python
students = {
"Rahul": {"Python", "AI"},
"Simran": {"Java", "Python"},
"Aman": {"DBMS"},
"Neha": {"Python", "DBMS"}
}
python_students = []
#Print names of students studying Python.
for name,subs in students.items():
    if "Python" in subs:
        python_students.append(name)
print(python_students)
```

```python
num = [10, 21, 32, 43, 54, 65]
grouped = {}
# Group Numbers by Remainder
grouped = {}

for n in num:
    rem = n % 3
    grouped.setdefault(rem, []).append(n)

for k in grouped:
    grouped[k] = tuple(grouped[k])
print(grouped)
```
