```python
import csv 
csv_filename = "OlympicMedals_2020.csv"
with open(csv_filename,encoding='utf-8',newline='') as csv_file:
    headers = csv_file.readline().strip('\n').split(',')
    print(f'Column headers : {headers}')
    reader = csv.reader(csv_file)
    for row in reader:
        print(row)
```

```python
import csv

with open("OlympicMedals_2020.csv", encoding="utf-8", newline="") as f:
    reader = csv.DictReader(f)

    for row in reader:
        print(row["Country"], row["Gold"])
```

```python
input_filename = 'country_info.txt'
with open(input_filename,encoding='utf-8',newline='') as countries_data:
    sample = countries_data.read()
    country_dialect = csv.Sniffer().sniff(sample)
    countries_data.seek(0)
    country_reader = csv.reader(
        countries_data,
        dialect = country_dialect
    )
    for row in country_reader:
        print(row)
```

```python
input_filename = 'country_info.txt'
with open(input_filename,encoding='utf-8',newline='') as countries_data:
    sample = ""
    for line in range(3):
        sample += countries_data.readline()
    country_dialect = csv.Sniffer().sniff(sample)
    countries_data.seek(0)
    country_reader = csv.reader(
        countries_data,
        dialect = country_dialect
    )
    for row in country_reader:
        print(row)
```
