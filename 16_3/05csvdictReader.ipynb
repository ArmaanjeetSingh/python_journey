```python
import csv

input_filename = 'country_info.txt'
countries = {}

dialect = csv.excel
dialect.delimiter = '|'

with open(input_filename,encoding='utf-8',newline='') as input_file:
    headings = input_file.readline().strip('\n').split(dialect.delimiter)
    reader = csv.DictReader(input_file,dialect = dialect,fieldnames = headings)
    for row in reader:
        countries[row['Country'].casefold()] = row
        countries[row['CC'].casefold()] = row
print(countries)
```

```python
#Write a program to count how many countries participated in the Olympics
count = 0
with open("OlympicMedals_2020.csv","r",encoding='utf-8',newline='') as medals_file:
    reader = csv.DictReader(medals_file)
    for row in reader:
        if row['Country']:
            count += 1
print(count)
```

```python
#sum of all Gold medals from the file.
total_golds = 0
with open("OlympicMedals_2020.csv","r",encoding='utf-8',newline='') as medals_file:
    reader = csv.DictReader(medals_file)
    for row in reader:
        total_golds += int(row["Gold"])
print(total_golds)
```

```python
#Find the country that won the most Gold medals.
max_gold = 0
country_name = ""
with open("OlympicMedals_2020.csv","r",encoding='utf-8',newline='') as medals_file:
    reader = csv.DictReader(medals_file)
    for row in reader:
        gold = int(row['Gold'])
        if max_gold < gold:
            max_gold = gold
            country_name = row['Country']
print(country_name,max_gold)
```

```python
new_filename = "top_countries.csv"
with open("OlympicMedals_2020.csv","r",encoding='utf-8',newline='') as medals_file, open(new_filename,"w",encoding='utf-8',newline='') as new_file  :
    headers = ['Country','Gold','Total']
    reader = csv.DictReader(medals_file)
    writer = csv.DictWriter(new_file,fieldnames=headers)
    writer.writeheader()
    for row in reader:
        if int(row['Total']) > 50:
            writer.writerow({
                'Country': row['Country'],
                'Gold': row['Gold'],
                'Total': row['Total']
            })


```
