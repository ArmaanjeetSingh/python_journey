```python
numbers = [10, 21, 32, 43, 54, 65]
grouped = {}
for num in numbers:
    rem = num % 3
    grouped.setdefault(rem,[]).append(num)
print(grouped)
```

```python
class Student:

    def __init__(self, name, marks):
        self._name = name
        self._marks = marks

    def get_name(self):
        return self._name
    
    def get_marks(self):
        return self._marks

    def set_marks(self,new_marks):
        self._marks  = new_marks


students = []

students.append(Student("Aman", 78))
students.append(Student("Riya", 91))
students.append(Student("Karan", 85))

# print students with marks > 80
for student in students:
    if student.get_marks() > 80:
        print(student.get_name())
```

```python
class BankAccount :
    def __init__(self,account_holder,balance):
        self._account_holder = account_holder
        self._balance =  balance

    def deposit(self,amount):
        if amount > 0:
            self._balance += amount
        else:
            print("Amount is negative")

    def withdraw(self,amount):
        if amount <= self._balance:
            self._balance -= amount
        else:
            print("Insufficient balance")

    def get_balance(self):
        return self._balance
acc = BankAccount("Vinayak", 1000)

acc.deposit(500)
acc.withdraw(200)

print(acc.get_balance())
```

```python
class Book:
    def __init__(self,title,author,available = True):
        self._title = title
        self._author = author
        self._available = available

    def borrow_book(self):
        if self._available == True:
            self._available = False
            print(f"{self._title} is borrowed")
        else:
            print("Book not available")

    def return_book(self):
        self._available = True

    def get_title(self):
        return self._title
    def get_author(self):
        return self._author

class Library :
    def __init__(self):
        self._book_list = []

    def add_book(self,Book):
        self._book_list.append(Book)
        print("Book {} is added to library".format(Book.get_title()))

    def show_books(self):
        for book in self._book_list:
            print(f"{book.get_title()} by {book.get_author()}")

    def borrow_book(self,title):
        for book in self._book_list:
            if book.get_title() == title:
                book.borrow_book()

library = Library()

library.add_book(Book("Python", "Tim"))
library.add_book(Book("FastAPI", "Sebastian"))

library.show_books()

library.borrow_book("Python")
```

```python
class Product :
    def __init__(self,name,price):
        self._name = name
        self._price = price

    def get_name(self):
        return self._name
    def get_price(self):
        return self._price


class Cart :
    def __init__(self):
        self._product_list = []

    def add_product(self,Product):
        self._product_list.append(Product)
        print(f"{Product.get_name()} is added to cart")

    def total_price(self):
        total = 0
        for product in self._product_list:
            total += product.get_price()
        print("Total price for cart : ",total)

cart = Cart()

cart.add_product(Product("Laptop", 60000))
cart.add_product(Product("Mouse", 500))

cart.total_price()
```
