```python
from datetime import datetime,timezone

now_local = datetime.now()
now_utc = datetime.now(timezone.utc)

print("local : ",now_local)
print("UTC : ",now_utc)
```

```python
from datetime import  timedelta
now = datetime.now()

future = now + timedelta(days=5)
past = now - timedelta(days = 0,hours=2)
print(future)
print(past)
```

```python
today = datetime.now()
birthday = datetime(year = 2026,month = 3, day = 24)
diff = birthday - today
print(diff.days, "days left")
```

```python
from datetime import datetime 
from zoneinfo import ZoneInfo

India_time = datetime.now(ZoneInfo("Asia/Kolkata"))
uk_time = datetime.now(ZoneInfo("Europe/London"))

print("India:", India_time)
print("UK:", uk_time)
```

```python
from datetime import datetime, timezone, timedelta
from zoneinfo import ZoneInfo

ist = datetime.now(ZoneInfo("Asia/Kolkata"))
utc = ist.astimezone(timezone.utc)
utc += timedelta(days=5,hours=2)
ist = utc.astimezone(ZoneInfo("Asia/Kolkata"))

print("IST:", ist)
print("UTC:", utc)
```

```python
from datetime import datetime, timezone
from zoneinfo import ZoneInfoNotFoundError

utc_time = datetime.now(timezone.utc)
utc_time += timedelta(days=5,hours=3)
us_time = utc_time.astimezone(ZoneInfo("America/New_York"))

print("UTC:", utc_time)
print("US + 5:", us_time)
```

```python
#mini program , take time from user
from datetime import datetime, timezone
from zoneinfo import ZoneInfo

year = int(input("Enter year: "))
month = int(input("Enter month: "))
day = int(input("Enter day: "))
hour = int(input("Enter hour (0-23): "))
minute = int(input("Enter minute: "))

user_time = datetime(year, month, day, hour, minute, tzinfo=ZoneInfo("Asia/Kolkata"))
utc_time = user_time.astimezone(timezone.utc)

us_time = user_time.astimezone(ZoneInfo("America/New_York"))
print("IST Time:", user_time)
print("UTC Time:", utc_time)
print("US Time:", us_time)
```
