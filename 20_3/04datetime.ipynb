```python
from datetime import datetime

now = datetime.now()

print(now.strftime("%Y-%m-%d"))
print(now.strftime("%d/%m/%Y"))
print(now.strftime("%A"))
print(now.strftime("%I:%M:%p"))
```

```python
date_str = "20-03-2026 12:14"

dt = datetime.strptime(date_str,"%d-%m-%Y %H:%M")
print(dt)
```

```python
#Convert India time → US time
from datetime import datetime
from zoneinfo import ZoneInfo

india_time = datetime.now(ZoneInfo("Asia/Kolkata"))
us_time = india_time.astimezone(ZoneInfo("America/New_York"))

print("US : {}".format(datetime.strftime(us_time,"%d-%m-%Y , %I:%M%p")))

print("India : {}".format(datetime.strftime(india_time,"%d-%m-%Y , %I:%M%p")))
```

```python
from datetime import datetime, timedelta, timezone

now = datetime.now()

utc_time = now.astimezone(timezone.utc)
future = now + timedelta(days = 5)
past = now - timedelta(seconds = 7200)

print("Now:", now)
print("Future:", datetime.strftime(future,"%d-%m-%Y , %I:%M%p"))
print("Past:", past)
```

```python
from datetime import  datetime , timezone
from zoneinfo import ZoneInfo

utc_time = datetime.now(timezone.utc)
india_time = utc_time.astimezone(ZoneInfo('Asia/Kolkata'))
print("UTC:", datetime.strftime(utc_time,"%d-%m-%Y , %I:%M%p"))
print("India:", india_time)
```

```python
from datetime import datetime, timezone, timedelta
from zoneinfo import ZoneInfo 

utc_now = datetime.now(timezone.utc)
future = utc_now + timedelta(hours=2)
past = utc_now - timedelta(days=1)

uk_time = past.astimezone(ZoneInfo("Europe/London"))

print("Now:", utc_now)
print("Future:", future)
print("Past:", past)
print("UK past : ",datetime.strftime(uk_time,"%d-%m-%Y , %I:%M%p"))
```
