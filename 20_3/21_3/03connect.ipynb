```python
import sqlite3

conn = sqlite3.connect("music2.db")
cursor = conn.cursor()

cursor.execute("SELECT * FROM artists") #view all artists
print(cursor.fetchall())
```

```python
cursor.execute("SELECT name FROM artists")#specific columns
print(cursor.fetchall())
```

```python
cursor.execute("SELECT * FROM artists WHERE name = 'Adele'") #Filter data using Where
print(cursor.fetchall())
```

```python
cursor.execute("SELECT * FROM songs ORDER BY track DESC")#Sorting
print(cursor.fetchall())
```

```python
cursor.execute("SELECT * FROM songs ORDER BY track LIMIT 5")
print(cursor.fetchall())
```

```python
cursor.execute("SELECT COUNT(*) FROM songs") #Count records
print(cursor.fetchone())
```

```python
#Find albums of a specific artist
cursor.execute("""
SELECT a.name, al.name 
FROM artists AS a
INNER JOIN albums AS al
ON a._id = al.artist
""")
print(cursor.fetchall())
```

```python
cursor.execute(""" SELECT * FROM sqlite_schema """)
print(cursor.fetchall())
```

```python
cursor.execute("""
SELECT s.title,a.name FROM songs AS s
  INNER JOIN albums AS a
  ON a._id = s.album
""")
print(cursor.fetchall())
```

```python
cursor.execute("""
SELECT songs.title, albums.name
FROM songs
JOIN albums ON songs.album = albums._id
""")
print(cursor.fetchall())
```

```python
#full relation between song, album, artist
cursor.execute("""
SELECT s.title,s.track,a.name, at.name  FROM songs AS s
  INNER JOIN albums AS a ON a._id = s.album
  INNER JOIN artists As at ON at._id = a.name
""")
print(cursor.fetchall())
```
