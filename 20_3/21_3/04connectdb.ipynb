```python
import sqlite3
db = sqlite3.connect("music2.db")
cursor = db.cursor()
```

```python
cursor.execute("SELECT name from albums;")
print(cursor.fetchall())
```

```python
cursor.execute("SELECT title,track FROM songs WHERE track > 5")
print(cursor.fetchall())
```

```python
cursor.execute("SELECT name FROM artists ORDER BY _id ASC")
print(cursor.fetchall())
```

```python
#Count number of albums per artist
cursor.execute("SELECT artist, COUNT(*) FROM albums GROUP BY artist")
print(cursor.fetchall())
```

```python
#Find all songs of a specific album
cursor.execute("SELECT title FROM songs GROUP BY album = 1")
print(cursor.fetchall())
```

```python
#Show albums with their artist names
cursor.execute(""" 
SELECT albums.name, artists.name FROM albums
 JOIN artists ON artists._id  = albums.artist
""")
print(cursor.fetchall())
```

```python
#Find artist with most songs
cursor.execute("""
SELECT at.name,COUNT(s._id) AS no_of_songs FROM songs AS s
  INNER JOIN albums AS a ON a._id = s.album
  INNER JOIN artists AS at ON at._id = a.artist
  GROUP BY at._id ORDER BY no_of_songs DESC
""")
print(cursor.fetchone())
```

```python
#List all songs with artist name sorted by artist
cursor.execute("""
SELECT at.name,s.title FROM songs AS s
  INNER JOIN albums AS a ON a._id = s.album
  INNER JOIN artists AS at ON at._id = a.artist
  ORDER BY at.name
""")
print(cursor.fetchall())
```
