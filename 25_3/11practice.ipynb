```python
import os
import fnmatch
def find_hidden_files(root_dir):
    for path,dirs,files in os.walk(root_dir):
        for hidden in fnmatch.filter(files,'*.emp3'):
            yield os.path.join(path,hidden)



root = 'music'
hidden_gen = find_hidden_files(root)

for file_path in hidden_gen:
    print(f"Found hidden file: {file_path}")
```

```python
import os 
import fnmatch

def find_mp3(root):
    for path,_,files in os.walk(root):
        for file in files:
            if file.endswith(".emp3"):
                yield os.path.join(path,file)


for song in find_mp3("music"):
    print(song)
```

```python
import os 
import fnmatch

def find_artists(root,pattern):
    pattern = pattern.upper()

    for path,dirs,_ in os.walk(root):
        for d in dirs:
            if fnmatch.fnmatch(d.upper(),pattern):
                yield os.path.join(path,d)


for artist in find_artists("music", "blue*"):
    print(artist)
```

```python
#Count songs per album
import os 

def count_songs_per_album(root):
    for path,dirs,files in os.walk(root):
        if files:
            yield path,len(files)

for album, count in count_songs_per_album("music"):
    print(f'{album} has {count} songs')
```

```python
def largest_album():
    max_count = 0
    max_album = None
    for album,count in count_songs_per_album("music"):
        if count > max_count:
            max_count = count
            max_album = album
    return max_album,max_count
print(largest_album())
```

```python
import os 

def get_size_songs(root):
    for path,_,files  in os.walk(root):
        for file in files:
            full_path = os.path.join(path,file)
            yield full_path,os.path.getsize(full_path)/1024/1024

for song in get_size_songs("music"):
    print(song)
```

```python
import os 

def find_empty_folders(root):
    for path,dirs,files in os.walk(root):
        if not dirs and not files:
            yield path

for folder in find_empty_folders("music"):
    print(folder)        
```

```python
import os

def rename_all_songs(root):
    for path,dirs,files in os.walk(root):
        for file in files:
            old_path = os.path.join(path,file)
            new_name = file.lower()
            new_path = os.path.join(path,new_name)
            if old_path != new_path:
                if os.path.exists(new_path):
                    print(f'Renamed {old_path} to {new_path}')
                else:
                    print('Skipped (exists) {}'.format(new_name))

rename_all_songs('music')
```

```python
import os

def get_artist_with_max_albums(root):
    max_count = 0
    max_artist = None

    for artist in os.listdir(root):
        artist_path = os.path.join(root,artist)
        if os.path.isdir(artist_path):
            if max_count < len(os.listdir(artist_path)):
               max_artist = artist_path
               max_count = len(os.listdir(artist_path))
    return max_count,os.path.abspath(max_artist)

get_artist_with_max_albums('music/music')      
```
