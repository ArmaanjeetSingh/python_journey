```python
nums = [1,2,3,4,5,6,7,8]
squares = [item**2 for item in nums]
print(squares)
```

```python
words = ["apple", "bat", "car", "elephant", "dog"]
# Return words with length greater than 3
filtered_words = [word for word in words if len(word) > 3]
print(filtered_words)
```

```python
s = "programming"
#create set of unique vowels
unique_vowels = {ch for ch in s if ch in "aeiou"}
print(unique_vowels)
```

```python
matrix = [[1,2,3], [4,5,6], [7,8,9]]
#flatten a nested list
flatten_mat = [el for item in matrix for el in item]
print(flatten_mat)
```

```python
a = [1,2,3,4,5]
b = [4,5,6,7]
#find common elements.
common_element = {i for i in a if i in b}
print(common_element)
```

```python
nums = [1,2,2,3,4,3,5]
#remove duplicates but keep order
unique_nums= {i for i in nums}
print(unique_nums)
```

```python
a = [1,2,3]
b = ['a','b']
#cartesian product
cartesian_product1 = [(i,j)for i in a for j in b]
print(cartesian_product1)
```
