```python
nums = [1, 2, 3, 4, 5]

def square_num(num: int) -> int:
    return num ** 2

squared = list(map(square_num, nums))
print(squared)
```

```python
nums = [2, 4, 6, 8, 11]
#Check if any number is greater than 10
print(any(num > 10 for num in nums))
```

```python
nums = [1, 3, 5, -2]
#check for all numbers are positive
print(all(num > 0 for num in nums ))
```

```python
nums = [1, 2, 3, 4, 5, 6]
evens = list(filter(lambda x : x % 2 == 0,nums))
squared_evens = list(map(lambda x: x**2, evens))
print(squared_evens)
```

```python
nums = [1, 2, 3, 4]
import functools
#Find the product of all numbers
def multiply_val(x:int, y:int)->int:
    return x*y


product = functools.reduce(multiply_val, nums )
print(product)
```

```python
matrix = [[1,2,3], [4,5,6], [7,8,9]]
#flatten list
flatten_mat = [el for item in matrix for el in item]
print(flatten_mat)
```

```python
words = ["cat", "dog", "hello", "hello"]
#Check if a list contains any palindrome string
print(any(word for word in words if word[:]==word[::-1]))
```

```python
words = ["hello", "world", "Python"]
#Check if all strings are lowercase
print(all(word.islower() for word in words))
```

```python
users = [
    {"name": "A", "age": 17},
    {"name": "B", "age": 22},
    {"name": "C", "age": 19}
]
eligible_user = list(filter(lambda x : x['age'] > 18,users ))
print(eligible_user)
```

```python
fruits = ["apple", "banana", "cherry"]
fruits_upper = list(map(str.upper,fruits))
print(fruits_upper)
```

```python
nums = [1,5,3,0,4,9,2]
def find_max(x,y):
    return x if x > y else y
max_el = functools.reduce(find_max,nums)
print(max_el)
```

```python
nums = [1,2,3,4,5,6,7,8,9,15]
#return nums divisble by 3 and 5
comp_nums = [item for item in nums if item % 3 == 0 and item % 5 == 0]
print(comp_nums)
```

```python
nums = [i for i in range(1,31)]
comp_nums = [item for item in nums if item % 3 == 0 and item > 20]
print(comp_nums)
```

```python
l = ["I", "love", "Python"]
def concat_str(x:str,y:str)->str:
    return x + ' ' + y


concatenated_str = functools.reduce(concat_str,l)
print(concatenated_str)
```
