```python
data = [(1, 3), (4, 1), (2, 2)]
print(sorted(data,key = lambda x : x[1]))
```

```python
students = [
    {"name": "A", "marks": 70},
    {"name": "B", "marks": 85},
    {"name": "C", "marks": 60}
]
student = max(students,key = lambda x : x["marks"])
print(student)
```

```python
matrix = [[1, 2], [3, 4], [5, 6]]
flat_mat = [el**2 for item in matrix for el in item]
print(flat_mat)
```

```python
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flat_mat_evens = [el for item in matrix for el in item if el % 2== 0]
print(flat_mat_evens)
```

```python
nums = [1, 2, 3, 4]
nums_dict = {i:i**2 for i in nums}
print(nums_dict)
```

```python
nums = [10, 15, 20, 25, 30]
check_nums = [item for item in nums if item % 5 == 0 and item > 20]
print(check_nums)
```

```python
nums = [5, 12, 7, 18, 3, 20]
print(list(map(lambda x : x**2,((filter(lambda x : x > 10, nums))))))
```

```python
nums = [1, 2, 3, 4]
#using reduce function
import functools
squared_nums = []
squared_nums.append(functools.reduce(lambda x,y : y **2, nums))
print(squared_nums)
```

```python
matrix = [[1, 2], [13, 4], [10, 5]]
print(any(el > 10 for item in matrix for el in item))
```

```python
matrix = [[1, 2], [3, -1], [5, 6]]
print(all(el > 0 for item in matrix for el in item ))
```
