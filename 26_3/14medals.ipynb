```python
from medals_data import medals_table
for country in medals_table:
    country["total"] = country["gold"] + country["silver"] + country["bronze"]
    #     total = gold + silver + bronze
    # print(medal['country'],total)
      
```

```python
medals_with_total = [
    {**country,"total":country['gold']+country['silver']+country['bronze']}  for country in medals_table
]
print(medals_with_total)
```

```python
#Top 3 countries by total medals
print(sorted(medals_table,key=lambda x:x['total'],reverse=True)[:3])
```

```python
#filter countries with more than 20 golds and total medals than 60
top_countries = list(filter(lambda country : country['gold'] >= 20 and country['total'] >= 60,medals_table))
print(top_countries)
```

```python
#create list using map and lambda
countries_list = list(map(lambda x: f"{x['country']} - {x['total']}", medals_table))
print(countries_list)
```

```python
#Find total gold medals across all countries
import functools
total_gold = functools.reduce(lambda acc,x: acc + x['gold'], medals_table,0)
print(total_gold)
```

```python
# Find country with maximum bronze medals
max_bronze = functools.reduce(
    lambda acc, curr: curr if curr['bronze'] > acc['bronze'] else acc,
    medals_table
)
print(max_bronze)
```
