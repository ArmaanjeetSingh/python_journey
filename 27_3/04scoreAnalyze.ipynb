```python
students = [
    {"name": "Aman", "marks": 85},
    {"name": "Rahul", "marks": 40},
    {"name": "Sneha", "marks": 78},
    {"name": "Priya", "marks": 92},
    {"name": "Karan", "marks": 35}
]
```

```python
print(list(map(lambda x : x['marks'],students)))
```

```python
# Get students who passed
print(list(filter(lambda x: x['marks'],students)))
```

```python
# Add grace marks
print(list(map(lambda x : x.get('marks') +5 , students)))
```

```python
#Total marks of class
import functools
total_marks = functools.reduce(lambda acc,x: acc+x.get('marks'),students,0)
print(total_marks)
avg_marks = total_marks / len(students)
print(avg_marks)
```

```python
#Highest scorer
highest_scorer = functools.reduce(lambda x,y : y['marks'] if y['marks'] > x else x ,students,0)
print(highest_scorer)
```
