```python
orders = [
    {
        "id": 1,
        "customer": "Aman",
        "items": [
            {"name": "laptop", "price": 50000},
            {"name": "mouse", "price": 500}
        ]
    },
    {
        "id": 2,
        "customer": "Rahul",
        "items": [
            {"name": "keyboard", "price": 1500},
            {"name": "mouse", "price": 700}
        ]
    },
    {
        "id": 3,
        "customer": "Aman",
        "items": [
            {"name": "monitor", "price": 12000}
        ]
    }
]
```

```python
orders_dict = {}
for order in orders:
    print(order['customer'])
    for item in order['items']:
        orders_dict[order['customer']] = orders_dict.setdefault(order['customer'],0)+item['price']
print(orders_dict)

# customers_orders = {order['customer']:item['price'] for order in orders for item in order['items']}
# print(customers_orders)
```
