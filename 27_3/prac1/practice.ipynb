```python
def second_largest(nums):
    largest = float('-inf')
    second = float('-inf')
    for num in nums:
        if num > largest:
            second = largest
            largest = num
        elif num > second and num != largest:
            second = num
    return second
print(second_largest([10, 20, 4, 45, 99])) 
```

```python
def find_non_repeating(s):
    freq = {}
    for ch in s:
        freq[ch] = freq.get(ch,0)+1
    for ch in s:
        if freq[ch] == 1:
            return ch
    return None
find_non_repeating('aabccdeff')
```

```python
str1 =  "A man a plan a canal Panama"
def check_palindrome(str1):
    str1 = str1.split(' ')
    str1 = ''.join(str1)
    return str1[::].casefold() == str1[::-1].casefold()
check_palindrome(str1)
```

```python
l = [1,2,3,1,3,6,6]
seen = set()
duplicate = set()
for i in l:
    if i not in seen:
        seen.add(i)
    else:
        duplicate.add(i)

print(duplicate)
```

```python
a = [1,3,5]
b = [2,4,6]
result = []
while i < len(a) and j < len(b):
    if a[i] < b[j]:
        result.append(a[j])
        i += 1
    else:
        result.append(b[j])
        j += 1
result.extend(a[i:])
result.extend(b[j:])
print(result)
```
