```python
n = 10
for i in range(n+1):
    for j in range(n - i,0,-1):
        print(" ",end= " ")
    for k in range(2*i-1):
        print("*",end=' ')
    print()

```

```python
n = 10
for i in range(n+1)[::-1]:
    for j in range(n - i,0,-1):
        print(" ",end= " ")
    for k in range(2*i-1):
        print("*",end=' ')
    print()

```
