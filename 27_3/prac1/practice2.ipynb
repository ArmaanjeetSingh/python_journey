```python
n = 5
for i in range(0,n+1):
    for j in range(0,i):
        print("*",end='')
    print()
```

```python
n = 5
for i in range(1,n+1):
    for j in range(0,n-i):
        print(" ",end=' ')
    for k in range(0,i):
        print("*",end=' ')
    print()
```

```python
n = 5
for i in range(1,n+1):
    for j in range(n-i,0,-1):
        print(" ",end="")
    for k in range(i):
        print("*",end=' ')
    print()
```

```python
n = 5
for i in range(1,n+1)[::-1]:
    for j in range(n-i,0,-1):
        print(" ",end="")
    for k in range(i):
        print("*",end=' ')
    print()
```

```python
for i in range(1,n+1):
    for j in range(n-i,0,-1):
        print(" ",end="")
    for k in range(i):
        print("*",end=' ')
    print()
for i in range(1,n)[::-1]:
    for j in range(n-i,0,-1):
        print(" ",end="")
    for k in range(i):
        print("*",end=' ')
    print()
```

```python
n = 4
for i in range(1,n+1):
    for j in range(i):
        print("*",end='')

    for k in range(2*(n-i)):
        print(" ",end='')

    for j in range(i):
        print("*",end='')

    print()
for i in range(1,n)[::-1]:
    for j in range(i):
        print("*",end='')

    for k in range(2*(n-i)):
        print(" ",end='')

    for j in range(i):
        print("*",end='')

    print()
```
