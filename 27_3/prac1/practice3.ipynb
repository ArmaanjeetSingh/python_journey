```python
#factorial of number
def fact(num):
    if num == 0 or num == 1:
        return 1
    return num * fact(num-1)
print(fact(5))
```

```python
#simple interest using lambda
si = lambda p,t,r:(p*r*t)/100
p,t,r=8,6,7
res = si(p,t,r)
print(res)
```

```python
#check armstrong
def check_armstrong(num):
    temp = num
    temp1 = 0
    while num > 0:
        rem = num % 10
        num //= 10
        temp1 += rem ** 3
    print(temp1)
    return "armstrong" if temp1 == temp else "Not armstrong"
repr(check_armstrong(153))
```

```python
#check all prime numbers in interval
for i in range(1,50):
    if i == 2:
        print("2 is prime")
    elif i > 2:
        for j in range(2,i):
            if i % j == 0:
                break
                # print('composite')
        else:
        
            print(i,'is prime')
    else:
        print("Neither prime nor composite")
            
```

```python
arr = [10,324,45,90,2]
res = arr[0]
for i in range(1,len(arr)):
    if arr[i] > res:
        res = arr[i]
print(res)
```
