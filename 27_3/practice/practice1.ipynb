```python
#from operator import itemgetter
from operator import itemgetter
d = [
    {"name": "Nandini", "age": 20},
    {"name": "Manjeet", "age": 20},
    {"name": "Nikhil", "age": 19}
]
print("Soretd by age : ",sorted(d,key=itemgetter('age')))
print("Soretd by age : ",sorted(d,key=lambda x:(x['age'],x['name'])))
```

```python
votes = [ "john", "johnny", "jackie", "johnny", "john", "jackie", "jamie", "jamie", "john", "johnny", "jamie", "johnny", "john" ]
count_dict = {}
for item in votes:
    count_dict[item] = count_dict.setdefault(item,0)+1
print(max(sorted(count_dict), key=lambda x: count_dict[x]))
```

```python
#Remove all Duplicates Words from a given Sentence
input_sentence = "Geeks for Geeks"
input_list = input_sentence.split(' ')
input_set = ' '.join(list(dict.fromkeys(input_list)))
print(input_set)
```

```python
s1 = "Geeks for Geeks"
#List comprehension with set
a = s1.split()
seen = set()
res = [i for i in a if not(i in seen or seen.add(i))]
print(' '.join(res))
```
