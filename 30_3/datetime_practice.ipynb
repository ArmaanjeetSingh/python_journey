```python
from datetime import datetime,timezone
from zoneinfo import ZoneInfo
now_local = datetime.now()
local_utc = datetime.now(timezone.utc)
print(now_local.strftime('%d-%m-%Y'))
print(local_utc.strftime('%d-%m-%Y %I:%H:%S %p'))
```

```python
from datetime import datetime,timedelta
now_local = datetime.now()
future = now_local + timedelta(days = 5,minutes=5)
print(future)
```

```python
today = datetime.now()
birthday = datetime(minute=28,hour = 3,day= 4,month=8,year = 2005)
diff = today - birthday
print(diff/(365))
```

```python
from datetime import timezone,datetime
usa_time = datetime.now(ZoneInfo('America/New_York'))
print(usa_time)
ist_time = usa_time.astimezone(ZoneInfo('Asia/Kolkata'))
ist_time += timedelta(days=4)
utc_time = ist_time.astimezone(timezone.utc)
print(ist_time.strftime('%d-%m-%Y %I:%M %p'),utc_time.strftime('%d-%m-%Y %I:%M %p'))
```
