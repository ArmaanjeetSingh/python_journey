```python
import sqlite3

conn = sqlite3.connect("student.db")
cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS students (
    roll_no  INTEGER PRIMARY KEY,
    name TEXT,
    mark TEXT
)
 """)
```

```python
conn.commit()
```

```python
import json

marks = {"math": 78, "science": 54}
cursor.execute(""" 
INSERT INTO students (roll_no,name,mark) VALUES (?,?,?)
""",(121,'Armaan',json.dumps(marks)))
conn.commit()
```

```python
cursor.execute("DELETE FROM students WHERE roll_no = ?",(123,))
conn.commit()
```

```python
for row in cursor.execute("SELECT name,mark FROM students WHERE roll_no=?",(121,)):
    for sub,mark in json.loads(row[1]).items():
        print(sub,':',mark)
```

```python
for row in cursor.execute("SELECT name,mark FROM students WHERE roll_no=?",(121,)):
    name,marks = row
    print(name,json.loads(marks))
```

```python
conn.close()
```
