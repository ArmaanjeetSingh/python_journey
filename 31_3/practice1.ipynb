```python
l = [10, 20, 4, 45, 99, 99]
max_num = 0
second_num = -1
for item in l:
    if item > max_num:
       second_num = max_num
       max_num =item
    elif max_num > second_num and item != max_num:
        second_num = item
print(second_num)
```

```python
l = [1,2,2,3,4,3,5]
num = []
l.sort()
exists_num = 0
for i in l:
    if i != exists_num:
        exists_num = i
        num.append(exists_num)
    else:
        continue
       
print(num)
```

```python
s = "python is easy and python is powerful"
unique_set = set()
s = s.split(' ')
for i in s:
    if i not in unique_set:
        unique_set.add(i)
print(len(unique_set))
```

```python
d1 = {'a': 10, 'b': 20}
d2 = {'b': 5, 'c': 15}
d3 = {}
for k1,v1 in d1.items():
    for k2,v2 in d2.items():
        if k1 != k2:
            d3[k2] = v2
            d3[k1] = v1
        else:
            d3[k1] = v1 + v2
print(d3)
```

```python
# Write a function to check if a number is prime.
def check_prime(num:int)->bool:
    for i in range(2,num):
        if  num % i == 0:
            return 0
            break
    return 1
check_prime(24)
```
