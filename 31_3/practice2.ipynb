```python
students = {
    "A": 85,
    "B": 72,
    "C": 90,
    "D": 65
}
group_std = {}
#group students by mark range
for k,v in students.items():
    if 80 <= v <= 100:
        group_std.setdefault('80-100',[]).append(k)
    elif 60 <= v <= 79:
        group_std.setdefault('60-79',[]).append(k)
print(group_std)
```

```python
words = ["eat","tea","tan","ate","nat","bat"]
anagrams = {}
for word in words:
    key = ''.join(sorted(word))
    if key not in anagrams:
        anagrams[key] = []
    anagrams[key].append(word)

result = list(anagrams.values())

print(result)
```
