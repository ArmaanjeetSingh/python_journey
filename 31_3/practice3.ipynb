```python
l1 = [1, 3, 5]
l2 = [2, 4, 6]
l3 = []
while i < len(l1) and j < len(l2):
    if l1[i] < l2[j]:
        l3.append(l1[i])
        i+=1
    else:
        l3.append(l2[j])
        j+=1
l3.extend(l1[i:])
l3.extend(l2[j:])
print(l3)
```

```python
s1 = "listen"
s2 = "silent"

freq = {}

for ch in s1:
    freq[ch] = freq.get(ch, 0) + 1

for ch in s2:
    freq[ch] = freq.get(ch, 0) - 1

print(all(v == 0 for v in freq.values()))
```

```python
l = [1, 2, 3, 4, 5]
k = 6
res = []
for i in range(len(l)):
    for j in range(1,len(l)):
        if l[i] + l[j] == k and i!=j:
            if (l[j],l[i]) not in res:
                res.append((l[i],l[j]))
print(res)
```

```python
arr = [5, 3, 8, 4, 2]
n = len(arr)

for i in range(n):
    for j in range(0,n-i-1):
        if arr[j] < arr[j+1]:
            arr[j],arr[j+1] = arr[j+1],arr[j]

print(arr)
```

```python
l = [0,1,0,3,12]
count_zero = l.count(0)

for _ in range(count_zero):
    l.remove(0)
    l.append(0)
print(l)
```

```python
words = ["eat","tea","tan","ate","nat","bat"]
grouped = {}
for word in words:
    key = ''.join(sorted(word))
    if key not in grouped:
        grouped[key] = []
    grouped[key].append(word)
print(grouped.values())
```

```python
l = [1,[2,3],[4,[5,6]]]
def flatten(lst):
    res = []
    for item in lst:
        if isinstance(item,list):
            res.extend(flatten(item))
        else:
            res.append(item)
    return res

print(flatten(l))
```
