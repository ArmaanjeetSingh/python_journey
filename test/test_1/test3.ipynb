```python
def store_list(*args):
    l = []
    for i in args:
        l.extend(i)
    
    for i in l:
        if l.count(i) > 1:
            l.remove(i)


    return l
l = store_list([1,2],[3,4,5],[5,6,7])
print(l)
```

```python
def store_list(*args,end='*'):
   l = []
   for i in args:
      l.extend(i)

   result = []
   seen = set()

   for i in l:
      if i not in seen:
         result.append(i)
         seen.add(i)

   return result
    
l = store_list([1,2],[3,4,5],[5,6,6,6,7])
print(l)
```
